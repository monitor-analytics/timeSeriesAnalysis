```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import adfuller

# Stationary dataset
np.random.seed(42)  # For reproducibility
data_length = 100
time_series = np.random.normal(loc=0, scale=1, size=data_length)

# Perform Augmented Dickey-Fuller test
adf_result = adfuller(time_series)

# Plot the data
plt.figure(figsize=(10, 6))
plt.plot(time_series)
plt.title("Generated Stationary Dataset")
plt.xlabel("Time")
plt.ylabel("Value")
plt.grid(True)
plt.show()

# Output ADF test result
adf_result
```

### Non-stationary data

```python
# Non-stationary dataset with a trend and seasonality
time = np.arange(data_length)
trend = time * 0.1  # Linear trend
seasonality = 10 * np.sin(np.pi * time / 12)  # Seasonal component with a yearly cycle (12 months)

non_stationary_series = trend + seasonality + np.random.normal(loc=0, scale=2, size=data_length)

# Plot the non-stationary data
plt.figure(figsize=(10, 6))
plt.plot(non_stationary_series)
plt.title("Generated Non-Stationary Dataset")
plt.xlabel("Time")
plt.ylabel("Value")
plt.grid(True)
plt.show()

adfuller(non_stationary_series)
```
