```python
import numpy as np
import pandas as pd
from statsmodels.tsa.arima_process import ArmaProcess
from statsmodels.graphics.tsaplots import plot_pacf
import matplotlib.pyplot as plt

# Simulate a time series with AR(3) process
np.random.seed(42)  # For reproducibility
ar = np.array([1, -0.5, 0.25, -0.125])  # AR coefficients for an AR(3) model
ma = np.array([1])  # MA coefficient (not used here, just a placeholder)
AR_object = ArmaProcess(ar, ma)
simulated_data = AR_object.generate_sample(nsample=100)

# Convert the simulated data into a pandas Series (optional, for easier handling)
time_series = pd.Series(simulated_data)

# Plot the PACF
plt.figure(figsize=(10, 6))
plot_pacf(time_series, lags=20)
plt.show()
```
