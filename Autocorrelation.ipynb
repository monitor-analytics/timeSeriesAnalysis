```python
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf
```

```python
# Dataset with a positive autocorrelation
# Simulate AR(1) process: x_t = 0.7*x_{t-1} + e_t, where e_t is white noise
np.random.seed(42)
data_length = 100
e = np.random.normal(loc=0, scale=1, size=data_length)
x = np.zeros(data_length)
for t in range(1, data_length):
    x[t] = 0.7 * x[t-1] + e[t]

# Plot the positively autocorrelated time series
fig, axs = plt.subplots(figsize=(12, 6))
plt.plot(x)
plt.title("Generated Time Series with Positive Autocorrelation")
plt.xlabel("Time")
plt.ylabel("Value")

# Plot the autocorrelation function

plot_acf(x, lags=20, alpha=0.05, use_vlines=True, title="Autocorrelation Function (ACF)")
plt.tight_layout()
plt.show()
```

```python
# Dataset with negative autocorrelation
# Simulate AR(1) process: x_t = -0.5*x_{t-1} + e_t, where e_t is white noise
np.random.seed(42)
e = np.random.normal(loc=0, scale=1, size=data_length)
x_negative = np.zeros(data_length)
for t in range(1, data_length):
    x_negative[t] = -0.5 * x_negative[t-1] + e[t]

# Plot the negatively autocorrelated time series
plt.subplots(figsize=(12, 6))
plt.plot(x_negative)
plt.title("Generated Time Series with Negative Autocorrelation")
plt.xlabel("Time")
plt.ylabel("Value")

# Plot the autocorrelation function

plot_acf(x_negative, lags=20, alpha=0.05, use_vlines=True, title="Autocorrelation Function (ACF)")
plt.tight_layout()
plt.show()
```

```python
# No autocorrelation
np.random.seed(42)
white_noise = np.random.normal(loc=0, scale=1, size=data_length)

# Plot the white noise time series
plt.subplots(figsize=(12, 6))
plt.plot(white_noise)
plt.title("Generated White Noise Series")
plt.xlabel("Time")
plt.ylabel("Value")

# Plot the autocorrelation function
# plt.subplots(figsize=(12,6))
plot_acf(white_noise, lags=20, alpha=0.05, use_vlines=True, title="Autocorrelation Function (ACF)")
plt.tight_layout()
plt.show()
```
